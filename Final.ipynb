```python
import numpy as np
```

# Import data from S3

```python
df = pandas.read_excel('s3://shaw-ia241/COVID-19-geographic-disbtribution-worldwide.xlsx')
df[:10] #the top 10 rows
```

# Analysis

# 1. Total Global deaths per day

```python
sum_dot = df.groupby('dateRep').sum()['deaths']

sum_dot.plot()
```

# 2. Countries with the highest death toll

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
sum_death_by_country.nlargest(10).plot.bar()
```

# 3. Total cases in China per day

```python
china_data = df.groupby('dateRep').sum()['cases']
china_data.plot(x = 'dateRep', y = 'cases')
```

# 4. 15 countries with most cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']

sum_cases_by_country.nlargest(15).plot.pie()
```

# 5. USA deaths per month

```python
usa_data = df.loc[df['countriesAndTerritories'] == 'United_States_of_America']

usa_data.plot.scatter(x = 'month', y = 'deaths')
```
